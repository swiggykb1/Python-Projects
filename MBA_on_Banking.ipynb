```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('C:/Users/HP/Downloads/insurance_data.csv')
df
```

```python
df = df.drop_duplicates()
df
```

```python
df['State'].value_counts()
df
```

```python
basket_AP = (df[df['State'] == 'AP'].groupby(['InvoiceNo', 'Product'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket_AR = (df[df['State'] == 'AR'].groupby(['InvoiceNo', 'Product'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket_AS = (df[df['State'] == 'AS'].groupby(['InvoiceNo', 'Product'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket_BR = (df[df['State'] == 'BR'].groupby(['InvoiceNo', 'Product'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket_CG = (df[df['State'] == 'CG'].groupby(['InvoiceNo', 'Product'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket_GA = (df[df['State'] == 'GA'].groupby(['InvoiceNo', 'Product'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket_GJ = (df[df['State'] == 'GJ'].groupby(['InvoiceNo', 'Product'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket_HR = (df[df['State'] == 'HR'].groupby(['InvoiceNo', 'Product'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
def encode_units(x):
    if x<= 0:
        return 0
    if x >= 1:
        return 1

basket_encoded = basket_AP.map(encode_units)
basket_AP = basket_encoded

basket_encoded = basket_AR.map(encode_units)
basket_AR = basket_encoded

basket_encoded = basket_AS.map(encode_units)
basket_AS = basket_encoded


basket_encoded = basket_BR.map(encode_units)
basket_BR = basket_encoded

basket_encoded = basket_CG.map(encode_units)
basket_CG = basket_encoded

basket_encoded = basket_GA.map(encode_units)
basket_GA = basket_encoded

basket_encoded = basket_GJ.map(encode_units)
basket_GJ = basket_encoded

basket_encoded = basket_HR.map(encode_units)
basket_HR = basket_encoded
```

```python
#Apriori
frequent_itemsets = apriori(basket_AP, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules = rules.sort_values(['confidence', 'lift'], ascending = [False, False])
rules
```
