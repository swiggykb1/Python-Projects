```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_excel('C:/Users/HP/Downloads/Online Retail.xlsx')
df
```

```python
df['Description'] = df['Description'].str.strip()
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')]
df
```

```python
basket_G = (df[df['Country'] == 'Germany'].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket_F = (df[df['Country'] == 'France'].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket_UK = (df[df['Country'] == 'United Kingdom'].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket_Por = (df[df['Country'] == 'Portugal'].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
basket_Swe = (df[df['Country'] == 'Sweden'].groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
def encode_units(x):
    if x<= 0:
        return 0
    if x >= 1:
        return 1

basket_encoded = basket_G.map(encode_units)
basket_G = basket_encoded

basket_encoded = basket_F.map(encode_units)
basket_F = basket_encoded

basket_encoded = basket_UK.map(encode_units)
basket_UK = basket_encoded

basket_encoded = basket_Por.map(encode_units)
basket_Por = basket_encoded

basket_encoded = basket_Swe.map(encode_units)
basket_Swe = basket_encoded

basket_F.drop('POSTAGE', inplace=True, axis=1)
basket_G.drop('POSTAGE', inplace=True, axis=1)
basket_UK.drop('POSTAGE', inplace=True, axis=1)
basket_Por.drop('POSTAGE', inplace=True, axis=1)
basket_Swe.drop('POSTAGE', inplace=True, axis=1)
```

```python
#Apriori
frequent_itemsets = apriori(basket_F, min_support=0.07, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules = rules.sort_values(['confidence', 'lift'], ascending = [False, False])
rules
```

```python
#ECLAT
from apyori import apriori
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk',],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice Cream', 'Eggs']]
```

```python
freq_itemsets = pd.DataFrame(dataset)
freq_itemsets
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(df, min_support = 0.6, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='support', support_only=True, min_threshold=1)
rules
```

```python
rules = rules[['antecedents', 'consequents', 'support']]
rules
```
